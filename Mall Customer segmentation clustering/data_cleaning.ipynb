```python
rr = pd.read_csv('customer_data.csv')
rr
```

```python
ff = pd.read_csv('sales_data.csv')
ff
```

```python
merge = pd.merge(rr , ff , left_on='customer_id' , right_on ='customer_id' , how = 'inner')
```

```python
merge
```

```python
total_revenue = merge['quantity']* merge['price']
merge['total_revenue'] = total_revenue
merge
```

```python
age_group = merge['age'].apply(lambda x: 'young' if x< 30 else 'middle-aged' if x <50 else 'senior')
merge['age_group'] = age_group
merge
```

```python
month = pd.to_datetime(merge['invoice date']).dt.month
year = pd.to_datetime(merge['invoice date']).dt.year
merge['month']= month
merge['year']= year
merge
```

```python
merge.to_csv('mall_customer_data.csv' , index = False)
```
