```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('mall_customer_data.csv')
df
```

```python
plt.figure(figsize=(9,5))
sns.scatterplot(x='category', y='total_revenue', data=df)
plt.title('Total revenue by category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['total_revenue'], bins = 30 , kde=True)
plt.title('Distribution of Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.pie(df['age_group'].value_counts(), labels=df['age_group'].unique() , autopct='%1.1f%%')
plt.title('Age Group Distribution')
plt.show()
```

```python
mall_performance = df.groupby('shopping_mall')['total_revenue'].sum().sort_values(ascending=False)
mall_performance
```

```python
plt.figure(figsize=(12,6))
mall_performance.plot(kind='bar')
plt.title('Total Revenue by Shopping Mall')
plt.xlabel('Shopping Mall')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.show()
```
